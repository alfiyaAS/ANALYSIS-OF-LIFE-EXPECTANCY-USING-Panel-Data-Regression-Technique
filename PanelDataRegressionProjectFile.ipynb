```python
from sklearn.linear_model import LinearRegression
```

```python
Z_1 = df[['gdp_capita']]
Y_1 = df['lifeexpectancy']
lm3 = LinearRegression()
lm3.fit(Z_1,Y_1)
Ylm_3=lm3.predict(Z_1)
plt.figure(figsize=(20,8))
plt.scatter(Z_1, Ylm_3)
plt.xlabel("GDP PER CAPITA")
plt.ylabel("PREDICTED LIFEEXPECTANCY")
```

```python
Z_2= df[['co2mmt_percapita']]
Y = df['lifeexpectancy']
lm = LinearRegression()
lm.fit(Z_2,Y)
Ylm_4=lm.predict(Z_2)
plt.figure(figsize=(20,8))
plt.scatter(Z_2, Ylm_4)
plt.xlabel("CO2 MMT PER CAPITA")
plt.ylabel("PREDICTED LIFEEXPECTANCY")
```
